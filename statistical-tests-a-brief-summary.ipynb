#### Why we use statistical tests 🤔?

Statistical tests serve several important purposes in data analysis and scientific research. Overall, statistical tests play a critical role in making data-driven decisions, supporting scientific claims, and ensuring the reliability and validity of research findings. They help transform raw data into meaningful insights and conclusions.

- **Inference:** Statistical tests allow researchers to make inferences about populations based on sample data. They provide a structured way to draw conclusions and make statements about the likelihood of observed patterns or differences being due to random chance or actual effects.

- **Hypothesis Testing:** Hypothesis testing is a fundamental aspect of scientific research. Statistical tests help researchers test hypotheses by comparing observed data to expected outcomes under a null hypothesis. This process allows researchers to determine if there is enough evidence to reject the null hypothesis in favor of an alternative hypothesis.

- **Identify Patterns and Relationships:** Statistical tests help identify patterns, relationships, and associations in data. They allow researchers to determine whether the observed relationships between variables are statistically significant or likely to occur due to chance.

- **Quantify Effects:** Statistical tests provide a quantitative way to measure the strength and magnitude of effects. For example, they can determine the size of the difference between groups or the extent of association between variables.

- **Decision Making:** Statistical tests aid decision-making processes by providing objective and evidence-based information. They help researchers and decision-makers make informed choices about various options or strategies.

- **Comparison and Generalization:** Statistical tests allow for the comparison of different groups, conditions, or treatments. This enables researchers to generalize their findings to larger populations and draw broader conclusions.

- **Controlled Analysis:** Statistical tests provide a systematic way to control for confounding factors and isolate the effects of variables of interest. This is crucial for establishing causal relationships between variables.

- **Validation and Verification:** In fields such as quality control, statistical tests are used to validate processes, products, and systems. They help ensure that standards are met and that outcomes are consistent and reliable.

- **Model Building:** In predictive modeling and machine learning, statistical tests are used to select relevant features, evaluate model performance, and assess the significance of predictors.

- **Scientific Progress:** Statistical tests contribute to the advancement of knowledge by facilitating the comparison of research results across different studies and enabling the replication of findings.

# Chi-Square Test of Independence

The Chi-Square Test of Independence is a statistical test used to determine whether there is a significant association between two categorical variables. It helps to assess whether the observed frequency distribution of the two variables is different from what would be expected if they were independent of each other.

The Chi-Square Test of Independence has certain assumptions and conditions that need to be considered to ensure the validity and reliability of the test results. These assumptions include:

1. **Independence of Observations:** The observations in the contingency table should be independent of each other. This means that each case (row and column combination) should be unrelated to the others.

2. **Random Sampling:** The data used in constructing the contingency table should come from a random sample, and the sample size should be sufficiently large for the test to be valid.

3. **Expected Cell Frequencies:** The expected cell frequencies should be greater than 5 for most cells in the contingency table. Low expected cell frequencies can lead to inaccurate results.

4. **Categorical Variables:** The variables being analyzed must be categorical. If one or both of the variables are ordinal or continuous, the Chi-Square Test of Independence may not be appropriate. In such cases, other methods like logistic regression might be more suitable.

5. **Large Sample Size:** The Chi-Square test becomes more accurate as the sample size increases. For smaller sample sizes, consider using Fisher's Exact Test if the expected cell frequencies are too small.

6. **Non-Zero Cell Frequencies:** Each cell in the contingency table should have a non-zero frequency. An empty cell would not contribute to the analysis.

7. **Mutual Exclusivity:** The categories of the variables being compared should be mutually exclusive. Each case should belong to only one category in each variable.

8. **No Perfect Collinearity:** The variables should not be perfectly dependent on each other. Perfect collinearity can lead to difficulties in calculating the expected cell frequencies.

It's important to carefully consider these assumptions before applying the Chi-Square Test of Independence. Violations of these assumptions can affect the reliability of the test results. If any assumptions are not met, it's advisable to explore alternative methods or approaches.

```python
import numpy as np
from scipy.stats import chi2_contingency


# Generating random categorical data for two features
np.random.seed(42)
feature1 = np.random.choice(['A', 'B', 'C'], size=100)
feature2 = np.random.choice(['X', 'Y', 'Z'], size=100)

# Creating a contingency table
contingency_table = np.array([[np.sum((feature1 == 'A') & (feature2 == 'X')),
                               np.sum((feature1 == 'A') & (feature2 == 'Y')),
                               np.sum((feature1 == 'A') & (feature2 == 'Z'))],
                              [np.sum((feature1 == 'B') & (feature2 == 'X')),
                               np.sum((feature1 == 'B') & (feature2 == 'Y')),
                               np.sum((feature1 == 'B') & (feature2 == 'Z'))],
                              [np.sum((feature1 == 'C') & (feature2 == 'X')),
                               np.sum((feature1 == 'C') & (feature2 == 'Y')),
                               np.sum((feature1 == 'C') & (feature2 == 'Z'))]])

# Performing the Chi-Square Test of Independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Printing the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_val)
print("\nExpected Frequencies:")
print(expected)
```

The Chi-Square statistic value of 2.7417027417027415 that you obtained from running the Chi-Square Test of Independence represents the calculated test statistic based on your data. This statistic is a measure of how much the observed frequencies in your contingency table differ from the expected frequencies under the assumption of independence between the two categorical variables.

In the context of the Chi-Square Test of Independence:

- If the calculated Chi-Square statistic is small, it suggests that the observed frequencies are close to what would be expected under the assumption of independence. This could indicate that there is little evidence of a significant association between the two categorical variables.

- If the calculated Chi-Square statistic is large, it suggests that the observed frequencies differ significantly from what would be expected under the assumption of independence. This could indicate that there is evidence of a significant association between the two categorical variables.

To determine whether the observed differences are statistically significant, you need to compare the calculated Chi-Square statistic to a critical value from the Chi-Square distribution. This critical value depends on the chosen significance level (often denoted as alpha) and the degrees of freedom associated with your contingency table.

In your case, you also have the p-value associated with the Chi-Square statistic. The p-value tells you the probability of obtaining a Chi-Square statistic as extreme as the one calculated, assuming that the two categorical variables are actually independent. A low p-value (typically below the chosen significance level) indicates that the observed association is statistically significant, and you would reject the null hypothesis of independence.

In summary, the Chi-Square statistic value of 2.7417027417027415 provides information about the degree of association between the two categorical variables. To interpret this result fully, you should compare it to critical values from the Chi-Square distribution and consider the associated p-value to determine whether the association is statistically significant.

# Post-hoc tests between categorical features

After using Chi-square test of independence, performing post-hoc tests for categorical data analysis can be a bit more involved than in some other types of statistical tests. In the context of the Chi-Square Test of Independence, post-hoc tests aim to identify which specific categories are driving the significant association detected by the initial test. Here's a simplified approach using Python's scipy.stats and statsmodels libraries for post-hoc testing:

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
import statsmodels.api as sm

# Creating a contingency table with a significant association
contingency_table = np.array([[30, 15, 5],
                              [10, 25, 5],
                              [5, 10, 30]])

# Perform Chi-Square Test of Independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)


# Calculate adjusted residuals
residuals = sm.stats.Table(contingency_table).resid_pearson

# Set significance level (adjust as needed)
alpha = 0.05

# Identify significant cells based on adjusted residuals
significant_cells = np.argwhere(np.abs(residuals) > np.sqrt(2 * np.log(1/alpha)))


# Print significant cells and their corresponding categories
for elem in significant_cells:
    row_category = elem[0]
    col_category = elem[1]
    print(f"Significant association between '{row_category}' and '{col_category}'")
```

Great! Identifying a significant association between categories of two categorical features through post-hoc analysis is an important finding. This suggests that the two variables are not independent and that there is a meaningful relationship or association between them. # Cochran-Mantel-Haenszel (CMH) test

The Cochran-Mantel-Haenszel (CMH) test is used to assess the association between two categorical variables while controlling for a third categorical variable (stratification variable)

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generating random categorical data for three variables
np.random.seed(42)
feature1 = np.random.choice(['A', 'B', 'C'], size=100)
feature2 = np.random.choice(['X', 'Y', 'Z'], size=100)
stratification_variable = np.random.choice(['S1', 'S2', 'S3'], size=100)

# Create a DataFrame
data = pd.DataFrame({'Feature1': feature1, 'Feature2': feature2, 'Stratification': stratification_variable})

# Create contingency tables for each stratum
contingency_tables = []
strata = data['Stratification'].unique()
for stratum in strata:
    stratum_data = data[data['Stratification'] == stratum]
    contingency_table = pd.crosstab(stratum_data['Feature1'], stratum_data['Feature2'])
    contingency_tables.append(contingency_table)

# Perform Cochran-Mantel-Haenszel test
chi2_stat, p_val, dof, expected = chi2_contingency(np.stack(contingency_tables))

# Print the CMH test results
print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_val)
```

The Cochran-Mantel-Haenszel test assesses the association between two categorical variables while controlling for a third categorical variable (stratification variable). Let's interpret the results you've provided for the Cochran-Mantel-Haenszel test:

1. **Chi-Square Statistic:** The calculated Chi-Square statistic is 18.836. This value represents the overall discrepancy between the observed frequencies in the contingency tables across different strata and the expected frequencies under the assumption of no association.

2. **Degrees of Freedom:** The degrees of freedom (df) for the test are 20. In the context of the Cochran-Mantel-Haenszel test, the degrees of freedom are related to the number of strata and the dimensions of the contingency tables within each stratum.

3. **P-value:** The p-value is 0.533. This p-value indicates the probability of obtaining the observed Chi-Square statistic (or a more extreme value) under the assumption of no association between the categorical variables.

**Interpretation:**
Given that the p-value is relatively high (0.533), it suggests that there is not enough evidence to reject the null hypothesis of no association between the two categorical variables, after accounting for the stratification variable. In other words, the observed association between the variables across different strata is not statistically significant.

In practical terms, this indicates that the distribution of categories across the two features is not significantly different from what would be expected under the assumption of no association, even when considering different strata defined by the stratification variable.

It's important to note that a higher p-value (closer to 1) suggests weaker evidence against the null hypothesis. Always interpret the results in the context of your research question and domain knowledge. If the p-value is greater than your chosen significance level (e.g., 0.05), you might conclude that there is insufficient evidence to suggest a significant association after accounting for stratification.

# Cramér's V - Measure of Association

Cramér's V is a measure of association between two categorical variables. It quantifies the strength and direction of the relationship between the variables.

![](https://slideplayer.com/slide/14079512/86/images/7/Nominal+Measures%3A+Cramer%E2%80%99s+V+%28V%29.jpg)

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generating random categorical data for a contingency table
np.random.seed(42)
rows = ['A', 'B', 'C']
columns = ['X', 'Y', 'Z']

contingency_table = np.random.randint(10, 100, size=(len(rows), len(columns)))

# Perform Chi-Square Test of Independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Calculate Cramér's V
def cramers_v(chi2, n, min_dim):
    return np.sqrt(chi2 / (n * (min_dim - 1)))

n = np.sum(contingency_table)
min_dim = min(len(rows), len(columns))
cramers_v_value = cramers_v(chi2_stat, n, min_dim)

# Print Cramér's V value
print("Cramér's V:", cramers_v_value)
```

The value you've obtained for Cramér's V is 0.1534. Cramér's V is a measure of association between two categorical variables, and it ranges from 0 to 1. Here's how to interpret the value of Cramér's V:

- **0:** No association between the variables.
- **Close to 0:** Weak association.
- **Around 0.1-0.2:** Moderate association.
- **Around 0.3-0.4:** Relatively strong association.
- **Close to 1:** Very strong association.

In our case, with a Cramér's V value of 0.1534, the association between the categorical variables is moderate. This suggests that there is a notable but not extremely strong relationship between the variables in the contingency table you analyzed.

Keep in mind that the interpretation of Cramér's V should be considered alongside the context of your data and research question. It's also important to note that while Cramér's V provides a measure of association, it doesn't indicate the direction of the relationship between the variables.

# Contingency Coefficient

The Contingency Coefficient is another measure of association between categorical variables. It's calculated based on the Chi-Square statistic and takes into account the size of the contingency table. The Contingency Coefficient ranges from 0 to 1, where higher values indicate a stronger association between the variables. As with other measures of association, interpretation should consider the context and domain knowledge of the data being analyzed.

![](https://www.empirical-methods.hslu.ch/files/2017/02/calculating-pearsons-contingecy-coefficient-chi-square-contingency.png)

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generating random categorical data for a contingency table
np.random.seed(42)
rows = ['A', 'B', 'C']
columns = ['X', 'Y', 'Z']

contingency_table = np.random.randint(10, 100, size=(len(rows), len(columns)))

# Perform Chi-Square Test of Independence
chi2_stat, _, _, _ = chi2_contingency(contingency_table)

# Calculate Contingency Coefficient
def contingency_coefficient(chi2, n, min_dim):
    return np.sqrt(chi2 / (chi2 + n))

n = np.sum(contingency_table)
min_dim = min(len(rows), len(columns))
contingency_coefficient_value = contingency_coefficient(chi2_stat, n, min_dim)

# Print Contingency Coefficient value
print("Contingency Coefficient:", contingency_coefficient_value)
```

The value you've obtained for the Contingency Coefficient is 0.2120. The Contingency Coefficient is a measure of association between two categorical variables, and it ranges from 0 to 1. Here's how to interpret the value of the Contingency Coefficient:

- **0:** No association between the variables.
- **Close to 0:** Weak association.
- **Around 0.1-0.2:** Moderate association.
- **Around 0.3-0.4:** Relatively strong association.
- **Close to 1:** Very strong association.

In your case, with a Contingency Coefficient value of 0.2120, the association between the categorical variables is moderate. This suggests that there is a moderate level of relationship or association between the variables in the contingency table you analyzed.

Remember, the interpretation of the Contingency Coefficient should be considered within the context of your data and research question. Additionally, combining this measure with other measures and domain knowledge can provide a more comprehensive understanding of the association between the variables.

# Phi Coefficient



The Phi Coefficient is a measure of association between two binary categorical variables. It's based on the concept of the Chi-Square statistic and is specifically used for 2x2 contingency tables.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generating random binary categorical data for a 2x2 contingency table
np.random.seed(42)
variable1 = np.random.randint(0, 2, size=100)
variable2 = np.random.randint(0, 2, size=100)

# Create a contingency table
contingency_table = pd.crosstab(variable1, variable2)

# Perform Chi-Square Test of Independence
chi2_stat, _, _, _ = chi2_contingency(contingency_table)

# Calculate Phi Coefficient
def phi_coefficient(chi2, n):
    return np.sqrt(chi2 / n)

n = np.sum(contingency_table)
phi_coefficient_value = phi_coefficient(chi2_stat, n)

# Print Phi Coefficient value
print("Phi Coefficient:", phi_coefficient_value)
```

The Phi Coefficient ranges from 0 to 1, where higher values indicate a stronger association between the binary variables. Here's how to interpret the value of the Phi Coefficient:

- 0: No association between the binary variables.
- Close to 0: Weak association.
- Around 0.1-0.2: Moderate association.
- Around 0.3-0.4: Relatively strong association.
- Close to 1: Perfect association.

# McNemar's Test

McNemar's Test is a statistical test used to analyze paired nominal data. It's often used when you have two categorical variables with two levels each and want to determine whether there is a significant difference in proportions between the levels across the two variables.

![](https://www.ncbi.nlm.nih.gov/books/NBK560699/bin/McNemar__test.jpg)

```python
import numpy as np
import pandas as pd
from statsmodels.stats.contingency_tables import mcnemar

# Generating random binary categorical data for a paired test
np.random.seed(42)
before = np.random.randint(0, 2, size=100)
after = np.random.randint(0, 2, size=100)

# Create a DataFrame
data = pd.DataFrame({'Before': before, 'After': after})

# Create a 2x2 contingency table
contingency_table = pd.crosstab(data['Before'], data['After'])

# Perform McNemar's Test
result = mcnemar(contingency_table)

# Print McNemar's Test results
print("McNemar's Statistic:", result.statistic)
print("P-value:", result.pvalue)
```

Interpreting McNemar's Test:

- McNemar's statistic indicates the extent of agreement or disagreement between the two variables.
- The p-value indicates whether the observed changes between the 'Before' and 'After' conditions are statistically significant.

The results you obtained from McNemar's Test are as follows:

- McNemar's Statistic: 19.0
- P-value: 0.1189

Let's interpret these results:

- **McNemar's Statistic:** The calculated McNemar's statistic is 19.0. This statistic reflects the number of discordant pairs in your paired data. In other words, it represents the cases where the 'Before' and 'After' conditions do not agree.

- **P-value:** The p-value associated with McNemar's Test is 0.1189. This p-value indicates the probability of obtaining the observed McNemar's statistic (or a more extreme value) under the assumption of no difference between the 'Before' and 'After' conditions.

**Interpretation:**

# Logistic Regression

Logistic Regression is a statistical method used to model the relationship between one or more predictor variables (usually continuous or categorical) and a binary outcome variable. It's often used to predict the probability of the occurrence of a certain event or to classify cases into one of two classes.

Logistic regression, like any statistical model, relies on certain assumptions for its validity and meaningful interpretation. These assumptions ensure that the estimated coefficients and predictions are reliable. Here are the key assumptions of logistic regression:

1. **Binary Outcome:** Logistic regression assumes that the outcome variable is binary or categorical with two levels (e.g., 0 or 1, Yes or No).

2. **Independence of Observations:** Each observation should be independent of the others. This assumption is important to avoid issues with multicollinearity and biased coefficient estimates.

3. **Linearity of the Logit:** The relationship between the predictor variables and the log-odds of the outcome should be linear. This means that the log-odds should change linearly with changes in the predictor variables.

4. **No Multicollinearity:** Logistic regression assumes that there is no perfect multicollinearity, which occurs when predictor variables are highly correlated with each other. High multicollinearity can lead to unstable coefficient estimates and difficulty in interpretation.

5. **Large Sample Size:** While logistic regression is often robust to deviations from normality, larger sample sizes tend to improve the reliability of parameter estimates and hypothesis tests.

6. **No Outliers:** Outliers can have a disproportionate impact on the logistic regression model, particularly if they are influential points. It's advisable to check for influential points and outliers.

7. **No Endogeneity:** Endogeneity occurs when one or more predictor variables are correlated with the error term. This can lead to biased coefficient estimates and invalid inferences.

8. **Correct Specification:** The correct variables should be included in the model. Omitted variables or incorrect functional forms can lead to biased estimates and incorrect interpretations.

9. **No Perfect Separation:** Perfect separation occurs when a combination of predictor variables can predict the outcome perfectly, leading to infinite coefficient estimates. This often happens with small datasets or when there are redundant predictor variables.

10. **No Perfect Prediction:** Perfect prediction occurs when a predictor variable perfectly predicts the outcome. In this case, the model cannot be estimated and is not valid.

It's essential to check and address violations of these assumptions to ensure that your logistic regression model produces valid and reliable results. Diagnostic tools such as residual analysis, influential point detection, and model comparison can help assess these assumptions and guide model improvement.

# Independent Samples t-test

The Independent Samples t-test is a statistical test used to determine if there is a significant difference between the means of two independent groups. It's commonly used when you have two sets of data collected from different groups or conditions and you want to assess whether the means of the two groups are significantly different from each other.

The Independent Samples t-test has several assumptions that need to be considered to ensure the validity of the test results. These assumptions include:

1. **Independent Samples:** The observations in each group should be independent of each other. This means that the values in one group should not be related or dependent on the values in the other group.

2. **Normality:** The distribution of data within each group should be approximately normally distributed. This is especially important for small sample sizes (less than 30). Larger sample sizes tend to be more robust to deviations from normality.

3. **Homogeneity of Variance (Equal Variance):** The variances of the two groups should be roughly equal. This assumption is known as homogeneity of variance or homoscedasticity. Violations of this assumption can affect the accuracy of the t-test.

4. **Continuous Data:** The data should be continuous. The t-test assumes that the data is measured on a continuous scale, and the test may not be appropriate for ordinal or categorical data.

5. **Sample Size:** While the t-test is robust to deviations from normality for larger sample sizes, smaller sample sizes may require a more normal distribution.

6. **No Outliers:** Outliers can significantly affect the results of the t-test, especially if they are influential points that affect the group means or variances.

7. **Equal Sample Sizes:** The t-test assumes equal sample sizes in both groups. If the sample sizes are unequal, the Welch's t-test (an adaptation of the independent t-test) can be used.

We have to check these assumptions before applying the Independent Samples t-test. Violations of these assumptions can lead to inaccurate results and conclusions. You can assess these assumptions through visualizations (e.g., histograms, box plots, Q-Q plots), statistical tests (e.g., Shapiro-Wilk test for normality), and checking for homogeneity of variance (e.g., Levene's test). If assumptions are not met, consider using alternative methods or transformations of the data.

![](https://develve.net/files/T-test%202%20sample.png)

```python
import numpy as np
from scipy import stats

# Generating example data
np.random.seed(42)
group1_data = np.random.normal(60, 10, 50)  # Group 1 with mean 60 and standard deviation 10
group2_data = np.random.normal(65, 10, 60)  # Group 2 with mean 65 and standard deviation 10

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(group1_data, group2_data)

# Print the results
print("Independent Samples t-test Results:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
```

# One Way Anova

One-Way Analysis of Variance (ANOVA) is a statistical test used to compare the means of three or more groups to determine whether there are statistically significant differences among them. It's an extension of the Independent Samples t-test to more than two groups. One-Way ANOVA is used when you have a single categorical independent variable and a continuous dependent variable. The goal is to assess whether at least one group has a significantly different mean from the others.

**Assumptions of One-Way ANOVA:**

1. **Independence:** The observations within each group are independent of each other. This is similar to the assumption in t-tests.

2. **Normality:** The data within each group should be approximately normally distributed. Deviations from normality may affect the accuracy of the ANOVA results. Normality is especially important when the group sizes are small (less than 30).

3. **Homogeneity of Variance (Equal Variance):** The variances of the groups should be approximately equal. Violations of homogeneity of variance can lead to inaccurate results. There are adjustments like Welch's ANOVA or non-parametric alternatives if this assumption is not met.

4. **Mutually Exclusive Categories:** The groups being compared should be mutually exclusive, meaning that each observation belongs to only one group.

5. **Homogeneity of Regression Slopes (if applicable):** If you're conducting an ANOVA with a continuous predictor variable, the relationship between the predictor and the dependent variable should be the same across all groups.

One-Way ANOVA is used in various fields to compare means of multiple groups, such as in medical trials, education studies, and manufacturing quality control.

![](https://statstest.b-cdn.net/wp-content/uploads/2020/11/One-Way-ANOVA.jpg)

```python
import numpy as np
from scipy import stats

# Generating example data
np.random.seed(42)
group1_data = np.random.normal(60, 10, 50)  # Group 1 with mean 60 and standard deviation 10
group2_data = np.random.normal(65, 10, 60)  # Group 2 with mean 65 and standard deviation 10
group3_data = np.random.normal(70, 10, 70)  # Group 3 with mean 70 and standard deviation 10

# Combine all data into one array
all_data = np.concatenate([group1_data, group2_data, group3_data])


# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1_data, group2_data, group3_data)

# Print the results
print("One-Way ANOVA Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between at least two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
```

# Kruskal Wallis test

The Kruskal-Wallis test is a non-parametric statistical test used to compare the medians of three or more independent groups when the data does not meet the assumptions of normality or homogeneity of variance required for One-Way ANOVA. It's an extension of the Mann-Whitney U Test to more than two groups. Kruskal-Wallis is used when you have a single categorical independent variable and a continuous dependent variable. The test assesses whether there are significant differences in the distributions of the groups.

**Assumptions of Kruskal-Wallis Test:**

1. **Independence:** The observations within each group are independent of each other, similar to the assumption in other statistical tests.

2. **Random Sampling:** The data is collected through random sampling from each group.

3. **Homogeneous Shape:** The distribution of scores within each group should have a similar shape. While it's not an assumption of normality, groups with drastically different shapes might not yield reliable results.

The Kruskal-Wallis test is used in situations where the data is not normally distributed or when there are concerns about the assumptions of other tests. It's commonly used in fields such as social sciences, medicine, and environmental studies.

```python
import numpy as np
from scipy import stats

# Generating example data
np.random.seed(42)
group1_data = np.random.normal(60, 10, 50)  # Group 1 with mean 60 and standard deviation 10
group2_data = np.random.normal(65, 10, 60)  # Group 2 with mean 65 and standard deviation 10
group3_data = np.random.normal(70, 10, 70)  # Group 3 with mean 70 and standard deviation 10

# Combine all data into one array
all_data = np.concatenate([group1_data, group2_data, group3_data])

# Create group labels
group_labels = ['Group 1'] * len(group1_data) + ['Group 2'] * len(group2_data) + ['Group 3'] * len(group3_data)

# Perform Kruskal-Wallis test
h_statistic, p_value = stats.kruskal(group1_data, group2_data, group3_data)

# Print the results
print("Kruskal-Wallis Test Results:")
print("H-statistic:", h_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between at least two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
```

# Pearson's Correlation Coefficient

Pearson's correlation coefficient measures the strength and direction of a linear relationship between two continuous variables. 

Pearson's correlation coefficient is a widely used measure of the strength and direction of the linear relationship between two continuous variables. However, its validity depends on several assumptions. Violations of these assumptions can lead to inaccurate or misleading correlation values. The key assumptions for Pearson's correlation coefficient include:

1. **Linearity:** Pearson's correlation coefficient measures linear relationships. If the relationship between the variables is not linear, the correlation coefficient may not accurately capture the association.

2. **Independence:** The observations of the two variables should be independent of each other. Autocorrelation (correlation within one variable over time) can affect the results.

3. **Homoscedasticity:** The variability of the two variables should be roughly constant across all levels of the variables. If the variability changes with the level of the variables, it can affect the correlation.

4. **Normality:** The variables should be normally distributed. While Pearson's correlation is robust to deviations from normality for small-to-moderate sample sizes, extreme non-normality can influence results.

5. **No Outliers:** Outliers can disproportionately affect correlation values, especially when the sample size is small.

6. **Equality of Variances:** The variances of the two variables should be roughly equal. If the variances are substantially different, it can affect the correlation.

7. **Continuous Variables:** Pearson's correlation is appropriate for continuous variables. If one or both variables are categorical or ordinal, other correlation measures like Spearman's rank correlation or Kendall's tau-b should be considered.

8. **No Perfect Correlation:** If there is a perfect linear relationship between the variables (all data points lie on a straight line), the correlation coefficient may not be informative.

It's important to assess these assumptions when interpreting Pearson's correlation coefficients. Visual inspection of scatter plots, residual plots, and the normality of residuals can provide insights into these assumptions. Additionally, robustness checks and alternative correlation measures might be necessary if assumptions are not met.

Linear regression relies on several assumptions to ensure the validity and reliability of its results. It's important to assess these assumptions before interpreting the output of a linear regression model. Here are the key assumptions of linear regression:

1. **Linearity:** The relationship between the independent variable(s) and the dependent variable should be linear. This means that changes in the independent variable(s) should lead to proportional changes in the dependent variable.

2. **Independence of Residuals:** The residuals (the differences between actual and predicted values) should be independent of each other and not exhibit any systematic patterns. Autocorrelation in residuals can indicate violations of this assumption.

3. **Homoscedasticity:** The residuals should have constant variance across all levels of the independent variable(s). Heteroscedasticity (unequal variance) can lead to unreliable standard errors and biased parameter estimates.

4. **Normality of Residuals:** The residuals should be approximately normally distributed. Deviations from normality can affect the accuracy of hypothesis tests and confidence intervals associated with model parameters.

5. **No Multicollinearity:** If you have multiple independent variables, they should not be highly correlated with each other. Multicollinearity can make it difficult to interpret the individual effects of predictors.

6. **No Outliers or Influential Observations:** Outliers can unduly influence the regression line, affecting parameter estimates. Influential observations can heavily impact the fit of the model.

7. **No Endogeneity:** The independent variable(s) should be exogenous, meaning they are not influenced by the errors of the model. Endogeneity can lead to biased estimates.

8. **No Perfect Multicollinearity:** Perfect multicollinearity occurs when one independent variable is a perfect linear combination of others. This can lead to mathematical instability in the regression coefficients.

It's important to assess these assumptions before drawing conclusions from a linear regression model. You can use various diagnostic techniques and tools, such as scatter plots, residual plots, normality tests, variance inflation factor (VIF) analysis, and leverage plots, to evaluate the assumptions and address potential violations. If assumptions are not met, you might need to consider transformations, model adjustments, or alternative methods.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
from scipy import stats

# Generating example data
np.random.seed(42)
num_samples = 100

# Continuous predictor variables
continuous_feature1 = np.random.normal(50, 10, num_samples)
continuous_feature2 = np.random.normal(60, 15, num_samples)
continuous_feature3 = np.random.normal(70, 8, num_samples)

# Categorical predictor variable
categories = ['A', 'B', 'C']
categorical_feature = np.random.choice(categories, size=num_samples)

# Target variable
target_variable = 5 + 2 * continuous_feature1 + 3 * continuous_feature2 + 4 * continuous_feature3 + np.random.normal(0, 5, num_samples)

# Create a DataFrame
data = pd.DataFrame({'continuous_feature1': continuous_feature1,
                     'continuous_feature2': continuous_feature2,
                     'continuous_feature3': continuous_feature3,
                     'categorical_feature': categorical_feature,
                     'target_variable': target_variable})

# Encode categorical feature using one-hot encoding
data = pd.get_dummies(data, columns=['categorical_feature'], drop_first=True)

# Add a constant term for intercept in the regression
data = sm.add_constant(data)

# Fit linear regression model
X = data[['const', 'continuous_feature1', 'continuous_feature2', 'continuous_feature3', 'categorical_feature_B', 'categorical_feature_C']]
y = data['target_variable']
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())

# Residual plots
residuals = model.resid
fig = sm.qqplot(residuals, line='s')
fig.suptitle("Q-Q Plot of Residuals")

plt.tight_layout()
plt.show()

# Normality test (Shapiro-Wilk test)
shapiro_statistic, shapiro_pvalue = stats.shapiro(residuals)
print("\nShapiro-Wilk Test for Normality:")
print("Test statistic:", shapiro_statistic)
print("p-value:", shapiro_pvalue)

# Variance Inflation Factor (VIF) analysis
X_vif = X.drop('const', axis=1)
vif = pd.DataFrame()
vif["Feature"] = X_vif.columns
vif["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
print("\nVariance Inflation Factor (VIF) Analysis:")
print(vif)

# Leverage plot
sm.graphics.influence_plot(model, criterion="cooks")
plt.show()
```

# Nonlinear - Loglinear Analysis (numerical target and mix of numerical/categorical as predictors)

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder

# Generating example data
np.random.seed(42)
num_samples = 100

# Numerical features
numerical_features = {
    'height': np.random.uniform(150, 200, num_samples),
    'weight': np.random.uniform(50, 100, num_samples),
    'income': np.random.uniform(20000, 80000, num_samples)
}

# Categorical feature
categories = ['Category1', 'Category2', 'Category3']
categorical_feature = np.random.choice(categories, size=num_samples)

# Target variable
age = 25 + 0.5 * numerical_features['height'] + 0.3 * numerical_features['weight'] + 0.1 * numerical_features['income']
age += np.random.normal(0, 5, num_samples)  # Adding some noise

# Create a DataFrame
data = pd.DataFrame({**numerical_features, 'categorical': categorical_feature, 'age': age})

# Encode the categorical feature
label_encoder = LabelEncoder()
data['categorical_encoded'] = label_encoder.fit_transform(data['categorical'])

# Prepare data for linear regression
X = data[['height', 'weight', 'income', 'categorical_encoded']]
y = np.log(data['age'])

# Add a constant term to the predictor matrix for the intercept
X = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the regression summary
print(results.summary())
```

```python
import numpy as np
import pandas as pd
import pingouin as pg

# Generating example data
np.random.seed(42)
variable1 = np.random.normal(50, 10, 100)  # Continuous variable 1 with mean 50 and std deviation 10
variable2 = np.random.normal(60, 15, 100)  # Continuous variable 2 with mean 60 and std deviation 15
control_variable = np.random.normal(70, 12, 100)  # Control variable

# Create a DataFrame
data = pd.DataFrame({'Variable1': variable1,
                     'Variable2': variable2,
                     'ControlVariable': control_variable})

# Calculate partial correlation
partial_corr = pg.partial_corr(data=data, x='Variable1', y='Variable2', covar='ControlVariable')

# Print the results
print("Partial Correlation Results:")
print("Partial correlation coefficient:", partial_corr['r'].values[0])
print("p-value:", partial_corr['p-val'].values[0])
```

A p-value of 0.1929690559395653 from a partial correlation test indicates that the observed partial correlation coefficient is not statistically significant at the conventional significance level (such as 0.05). In other words, there isn't strong evidence to suggest that there's a significant linear relationship between the two continuous variables (variable1 and variable2), while controlling for the effect of the control variable.

When interpreting the p-value in the context of hypothesis testing:

- If the p-value is less than or equal to the chosen significance level (e.g., 0.05), you would typically reject the null hypothesis and conclude that there is a significant partial correlation between the variables of interest.
- If the p-value is greater than the chosen significance level, you would fail to reject the null hypothesis and conclude that there is no significant partial correlation between the variables of interest.

In the case, since the p-value is 0.1929690559395653, which is greater than 0.05, you would likely fail to reject the null hypothesis and conclude that there is no significant partial correlation between variable1 and variable2, while controlling for the effect of the control variable. This means that any observed correlation could be due to chance, and there might not be a meaningful relationship between the variables once the control variable is taken into account.

# Robust Regression

Robust regression is used when there are concerns about the assumptions of traditional linear regression, such as the presence of outliers or heteroscedasticity. One common method of robust regression is the Huber loss, which is less sensitive to outliers compared to the least squares loss.

Robust regression is a type of regression analysis that aims to provide reliable parameter estimates even when some assumptions of classical linear regression are violated. While robust regression methods are designed to be more resistant to outliers and violations of normality and homoscedasticity, they still have certain assumptions. Here are the main assumptions of robust regression:

1. **Linearity:** Like classical linear regression, robust regression assumes that the relationship between the predictor variables and the response variable is linear. This assumption ensures that the model accurately represents the underlying relationships in the data.

2. **Independence:** The observations should be independent of each other, meaning that the errors of one observation should not depend on the errors of other observations. Violations of independence can lead to biased coefficient estimates.

3. **Homoscedasticity (Constant Variance):** While robust regression methods are more robust to violations of homoscedasticity, it's still beneficial to have roughly constant variance across all levels of the predictor variables.

4. **No Perfect Multicollinearity:** The predictor variables should not be perfectly correlated with each other. While robust regression can handle multicollinearity better than ordinary least squares, extremely high multicollinearity can still lead to unstable results.

5. **Outliers:** Robust regression methods are designed to be less influenced by outliers, but extreme outliers can still have a significant impact on the results. It's important to identify and understand the impact of outliers on your analysis.

6. **No Endogeneity:** The predictor variables should not be correlated with the error term. Endogeneity can lead to biased coefficient estimates and incorrect inferences.

It's important to note that while robust regression can handle violations of assumptions better than classical linear regression, there is no one-size-fits-all robust regression method that works well in all situations. Different robust regression methods have different strengths and weaknesses. Therefore, it's essential to understand your data and the specific assumptions being violated in order to choose an appropriate robust regression method and interpret the results accurately



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Generating example data
np.random.seed(42)
num_samples = 100

feature1 = np.random.normal(50, 10, num_samples)
feature2 = np.random.normal(60, 15, num_samples)
categorical_feature = np.random.choice(['A', 'B', 'C'], size=num_samples)
target = 2 * feature1 + 3 * feature2 + np.random.normal(0, 10, num_samples)

# Create a DataFrame
data = pd.DataFrame({'Feature1': feature1,
                     'Feature2': feature2,
                     'CategoricalFeature': categorical_feature,
                     'Target': target})

# Convert categorical feature to dummy variables
data = pd.get_dummies(data, columns=['CategoricalFeature'], drop_first=True)

# Separate features and target
X = data[['Feature1', 'Feature2', 'CategoricalFeature_B', 'CategoricalFeature_C']]
y = data['Target']

# Add constant to features for intercept
X = sm.add_constant(X)

# Fit robust regression model
model = sm.RLM(y, X, M=sm.robust.norms.HuberT())  # Huber loss for robustness
results = model.fit()

# Print model summary
print(results.summary())
```
